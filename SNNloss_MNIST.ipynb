### ***Initialization***

```python
import numpy as np
np.random.seed(0)
import tensorflow as tf
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)
tf.keras.backend.set_session(sess)

# Import MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
```

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
plt.style.use('ggplot')

def dim_reduction(X):
    X_embedded = TSNE(n_components=2, verbose=0).fit_transform(X)
    return X_embedded
    
def scatter_plot_with_label(layers,labels,title):
    num_plots = len(layers)
    plt.figure(figsize=(5*num_plots, 5))
    for l, X in enumerate(layers):
        if X.shape[1] != 2:
            # dimension reduction using t-SNE
            X = dim_reduction(X)
        
        plt.subplot(1,num_plots,l+1)
        plt.scatter(X[:,0], X[:,1], c=labels, cmap="terrain",alpha=1, marker="^", s=50)
        plt.title("Features in layer {}".format(l))
        plt.axis("on")
    #plt.title(title)
    plt.show()
```

```python
# In this example, we limit mnist data
X_train, Y_train = mnist.train.next_batch(20000)
X_test , Y_test  = mnist.test.next_batch(300) # small sampling for visualization
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

### ***Model Initialization***
- 4 Layer Fully Connected Layer
- 10 classifier with softmax

```python
tag = "MLP"
layers = []
# Simple model
with tf.name_scope('{}-placeholders'.format(tag)):
    x   = tf.placeholder(tf.float32, [None,784])
    y   = tf.placeholder(tf.float32, [None,10])

with tf.name_scope('{}-layer-1'.format(tag)):
    x1 = tf.layers.dense(inputs=x, units=300)
    #x1 = tf.layers.batch_normalization(x1)
    x1 = tf.nn.relu(x1)
    
print(x1)
layers.append(x1)

with tf.name_scope('{}-layer-2'.format(tag)):
    x2 = tf.layers.dense(inputs=x1, units=100)
    #x2 = tf.layers.batch_normalization(x2)
    x2 = tf.nn.relu(x2)
    
print(x2)
layers.append(x2)

with tf.name_scope('{}-layer-3'.format(tag)):
    x3 = tf.layers.dense(inputs=x2, units=50)
    #x3 = tf.layers.batch_normalization(x3)
    x3 = tf.nn.relu(x3)
    
print(x3)
layers.append(x3)

with tf.name_scope('{}-layer-4'.format(tag)):
    x4 = tf.layers.dense(inputs=x3, units=10)
    #x4 = tf.layers.batch_normalization(x4)
    p = x4
print(x4)
layers.append(x4)

with tf.name_scope('{}-layer-out'.format(tag)):
    p_act  = tf.nn.softmax(p)
```

### ***Soft Nearest Neighbors(SNN) loss function***
#### (1) Calculate pairwise distances

```python
STABILITY_EPS = 0.00001  # used to make the calculation of SNNL more stable
```

```python
def pairwise_cos_distance(A, B):
    """Pairwise cosine distance between two matrices.
    :param A: a matrix.
    :param B: a matrix.
    :returns: A tensor for the pairwise cosine between A and B.
    """
    normalized_A = tf.nn.l2_normalize(A, axis =1)
    normalized_B = tf.nn.l2_normalize(B, axis =1)
    prod = tf.matmul(normalized_A, normalized_B, adjoint_b=True)
    return 1 - prod
```

#### (2) Exponentiated pairwise distance between each element of A and all those of B.

```python
def expoentiated_distance(A, B, temperature):
    """Exponentiated pairwise distance between each element of A and
    all those of B.
    :param A: a matrix.
    :param B: a matrix.
    :param temp: Temperature
    :returns: A tensor for the exponentiated pairwise distance between
    each element and A and all those of B.
    """
    distance_matrix = pairwise_cos_distance(A, B)
    return tf.exp(-(distance_matrix / temperature))
```

#### (3)-1 Row normalized exponentiated pairwise distance between all the elements of x

```python
def pick_probability(x, temperature):
    """Row normalized exponentiated pairwise distance between all the elements
    of x. Conceptualized as the probability of sampling a neighbor point for
    every element of x, proportional to the distance between the points.
    :param x: a matrix
    :param temp: Temperature
    :returns: A tensor for the row normalized exponentiated pairwise distance
              between all the elements of x.
    """
    f   = expoentiated_distance(x, x, temperature) - tf.eye(tf.shape(x)[0])
    ret = f / (STABILITY_EPS + tf.expand_dims(tf.reduce_sum(f, 1), 1))
    return ret
```

#### (3)-2. Masking matrix such that element i,j is 1 iff y[i] == y2[i]

```python
def same_label_mask(y, y2):
    """Masking matrix such that element i,j is 1 iff y[i] == y2[i].
    :param y: a list of labels
    :param y2: a list of labels
    :returns: A tensor for the masking matrix.
    """
    return tf.cast(tf.squeeze(tf.equal(y, tf.expand_dims(y2, 1))), tf.float32)
```

#### (4). The pairwise sampling probabilities for the elements of x for neighbor points which share labels

```python
def masked_pick_probability(x, y, temperature):
    """The pairwise sampling probabilities for the elements of x for neighbor
    points which share labels.
    :param x: a matrix
    :param y: a list of labels for each element of x
    :param temp: Temperature
    :returns: A tensor for the pairwise sampling probabilities.
    """
    return pick_probability(x, temperature) * same_label_mask(y, y)
```

#### (5)-1. Soft Nearest Neighbor loss

```python
def SNNL(x, y, temperature):
    """Soft Nearest Neighbor Loss
    :param x: a matrix.
    :param y: a list of labels for each element of x.
    :param temp: Temperature.
    :returns: A tensor for the Soft Nearest Neighbor Loss of the points
              in x with labels y.
    """
    summed_masked_pick_prob = tf.reduce_sum(masked_pick_probability(x, y, temperature), 1)
    log_prob = -tf.log(STABILITY_EPS + summed_masked_pick_prob)
    ret = tf.reduce_mean(log_prob)
    return ret
```

#### (5)-2. Optimized Temperature Soft Nearest Neighbor loss

```python
def optimized_temp_SNNL(x, y, initial_temp):
    """The optimized variant of Soft Nearest Neighbor Loss. Every time this
    tensor is evaluated, the temperature is optimized to minimize the loss
    value, this results in more numerically stable calculations of the SNNL.
    :param x: a matrix.
    :param y: a list of labels for each element of x.
    :param initial_temp: Temperature.
    :returns: A tensor for the Soft Nearest Neighbor Loss of the points
              in x with labels y, optimized for temperature.
    """
    t = tf.Variable(1, dtype=tf.float32, trainable=False, name="temp")

    def inverse_temp(t):
        # pylint: disable=missing-docstring
        # we use inverse_temp because it was observed to be more stable when optimizing.
        return tf.div(initial_temp, t)

    ent_loss = SNNL(x, y, inverse_temp(t))
    updated_t = tf.assign(t, tf.subtract(t, 0.1*tf.gradients(ent_loss, t)[0]))
    inverse_t = inverse_temp(updated_t)
    return SNNL(x, y, inverse_t)
```

### ***Cross-entropy + layerwise-Soft Nearest Neighbors(SNN) = Entire loss***

```python
# Base loss - cross entropy
ce_loss = tf.losses.softmax_cross_entropy(logits=p, onehot_labels=y)

# loss as regulizer - SNNL
#factor: The balance factor between SNNL and ross Entropy. If factor is
#             negative, then SNNL will be maximized.
factor = 10.
temperature = 20.

loss_fn = optimized_temp_SNNL
#loss_fn = SNNL

for l in layers:
    print(l)
    print(tf.layers.flatten(l))

layerwise_losses = [loss_fn(tf.layers.flatten(layer), tf.argmax(y,axis=1), temperature) for layer in layers]
layerwise_losses = factor * tf.add_n(layerwise_losses)

# Total
loss = ce_loss + layerwise_losses
#loss = layerwise_losses

optimizer = tf.train.AdamOptimizer(1e-4).minimize(loss)
correct_pred = tf.equal(tf.argmax(p_act,1), tf.argmax(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

### ***Training - Cross-entropy + SNN loss***

```python
tf.global_variables_initializer().run(session=sess)
```

```python
epochs = 100
batch_size = 2000
for e in range(epochs):
    train_loss = []
    train_acc = []
    train_snn = []
    for i in range(0,len(X_train),batch_size):
        feed_dict = {x : X_train[i:i+batch_size],
                     y : Y_train[i:i+batch_size]}
        
        _ , batch_loss, batch_acc, batch_snn = sess.run([optimizer, 
                                                        loss, 
                                                        accuracy, 
                                                        layerwise_losses], 
                                                       feed_dict=feed_dict)
        train_loss.append(batch_loss)
        train_acc.append(batch_acc)
        train_snn.append(batch_snn)
        
    feed_dict = {x: X_test, y: Y_test}
    test_loss, test_acc, test_snn = sess.run([loss, accuracy, layerwise_losses], feed_dict=feed_dict)
    test_preds = sess.run(layers, feed_dict=feed_dict)
    
    if e % 10 == 0:
        scatter_plot_with_label(test_preds, np.argmax(Y_test,axis=1), title="Feature distribution of each layer used snn loss")     
    if e % 1 == 0:
        print("epoch [{:>4}] - loss : {:3.4f}, acc : {:3.4f}, snn : {:3.4f} - test loss : {:3.4f}, test acc : {:3.4f}, test snn : {:3.4f}".format(e, 
                                                                                                                                                 np.mean(train_loss), 
                                                                                                                                                 np.mean(train_acc),
                                                                                                                                                 np.mean(train_snn),
                                                                                                                                                 test_loss,
                                                                                                                                                 test_acc,
                                                                                                                                                 test_snn))
```
